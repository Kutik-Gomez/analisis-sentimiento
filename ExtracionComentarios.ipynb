```python
import pandas as pd
import os

# Reemplaza 'ruta/al/archivo.csv' con la ruta a tu archivo CSV
archivo_csv = './Dataset-IEESQuito/14 UNIDAD  DE ATENCIÓN AMBULATORIA  SANGOLQUI.csv'

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)
df
```

```python
# Reemplaza 'nombre_columna' con el nombre de la columna que deseas extraer
comentario = 'wiI7pd'  # Especifica el nombre de la columna que quieres extraer

# Seleccionar la columna deseada
columna_extraida = df[[comentario]]  # Se usa doble corchete para mantener el formato DataFrame

# Reemplaza 'nuevo_archivo.csv' con el nombre del nuevo archivo CSV
nuevo_archivo_csv = 'comentarios14.csv'
nueva_ruta = f'./Dataset-Comentarios/{nuevo_archivo_csv}'

# Guardar la columna extraída en un nuevo archivo CSV
columna_extraida.to_csv(nueva_ruta, index=False)

print(f"Columna '{comentario}' extraída y guardada en '{nueva_ruta}'")
```
